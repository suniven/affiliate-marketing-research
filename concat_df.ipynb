```python
excluded_extensions = ['.js', '.pdf', '.svg','.jpg', '.css', '.png', '.jpeg', '.gif', '.ico', '.xml', '.json', '.mp3', '.wav', '.mp4', '.avi', '.mov', '.zip', '.rar']
```

```python
df1 = df[~df['original_url'].str.endswith(tuple(excluded_extensions))]
```

```python
df1.reset_index(inplace=True, drop=True)
```

```python
df1
```

```python
# 定义函数，统计分割后数组的最后一个元素出现的次数
def count_last_element(url):
    elements = url.split('.')
    last_element = elements[-1]
    return last_element.split('?')[0]

# 将函数应用到 url 列的每个元素上，并生成新的列 last_element
df1['last_element'] = df1['original_url'].apply(count_last_element)

# 统计 last_element 出现的次数，并按照出现次数从高到低排序
counts = df1['last_element'].value_counts().sort_values(ascending=False)
```

```python
df1.head()
```

```python
counts.head(40)
```

```python
filter_list=['jpg', 'org/2000/svg', 'js', 'com/webapp/www/pin/', 'png', 'css', 'jpeg', 'svg', 'ico']
```

```python
df1 = df1[~df1['last_element'].isin(filter_list)]
df1
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
# 定义函数，统计分割后数组的最后一个元素出现的次数
def get_domain(url):
    subdomain = url.split('/')[2]
    return '.'.join(subdomain.split('.')[-2:])

# 将函数应用到 url 列的每个元素上，并生成新的列 last_element
df1['domain'] = df1['original_url'].apply(get_domain)

# 统计 last_element 出现的次数，并按照出现次数从高到低排序
counts = df1['domain'].value_counts().sort_values(ascending=False)
```

Tianyou YouTube search 中的 URL 以及它们的重定向链接

```python
import pandas as pd
import os
```

```python
df = pd.read_csv('./data/tianyou/youtube/4.csv', engine='python')
```

```python
file_path = './data/tianyou/youtube/'
```

```python
files = os.listdir(file_path)
files
```

```python
df = pd.DataFrame()
df
```

```python
for file in files:
    print(file)
    df_tmp = pd.read_csv(file_path+file, engine='python')
    df = pd.concat([df, df_tmp])
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.at[0,'description']
```

```python
import re
url_regex = r'https?:\/\/[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]'
extracted_urls = re.findall(url_regex, df.at[0,'description'], re.MULTILINE)
extracted_urls
```

```python
import re
from tqdm import tqdm

url_regex = r'https?:\/\/[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]'

data = []
for i in tqdm(range(len(df))):
    if not pd.isna(df.at[i, 'description']):
        extracted_urls = re.findall(url_regex, df.at[i,'description'], re.MULTILINE)
        for url in extracted_urls:
            data.append({'original_url': url, 'source': df.at[i,'videoId']})
```

```python
df1 = pd.DataFrame(data)
df1.head()
```

```python
df1.shape
```

```python
df1.to_csv('./data/urls-without-redirection/youtube-search.csv', index = False)
```
