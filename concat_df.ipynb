Tianyou YouTube search 中的 URL 以及它们的重定向链接

```python
import pandas as pd
import os
```

```python
df = pd.read_csv('./data/tianyou/youtube/4.csv', engine='python')
```

```python
file_path = './data/tianyou/youtube/'
```

```python
files = os.listdir(file_path)
files
```

```python
df = pd.DataFrame()
df
```

```python
for file in files:
    print(file)
    df_tmp = pd.read_csv(file_path+file, engine='python')
    df = pd.concat([df, df_tmp])
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.at[0,'description']
```

```python
import re
url_regex = r'https?:\/\/[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]'
extracted_urls = re.findall(url_regex, df.at[0,'description'], re.MULTILINE)
extracted_urls
```

```python
import re
from tqdm import tqdm

url_regex = r'https?:\/\/[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]'

data = []
for i in tqdm(range(len(df))):
    if not pd.isna(df.at[i, 'description']):
        extracted_urls = re.findall(url_regex, df.at[i,'description'], re.MULTILINE)
        for url in extracted_urls:
            data.append({'original_url': url, 'source': df.at[i,'videoId']})
```

```python
df1 = pd.DataFrame(data)
df1.head()
```

```python
df1.shape
```

```python
df1.to_csv('./data/urls-without-redirection/youtube-search.csv', index = False)
```

```python
urls = list(set(df1.original_url.to_list()))
```

```python
len(urls)
```

```python
df2 = pd.DataFrame(columns=['original_url'], data=urls)
df2.head()
```

```python
df2.to_csv('./data/urls-to-be-redirected/youtube.csv', index=False)
```
