Tianyou Twitter search 中的 URL 以及它们的重定向链接，加预处理

```python
import pandas as pd
import os
```

```python
file_path = './data/tianyou/twitter/'
```

```python
files = os.listdir(file_path)
files
```

```python
df = pd.DataFrame()
df
```

```python
for file in files:
    df_tmp = pd.read_csv(file_path+file)
    df = pd.concat([df, df_tmp])
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
type(df.at[0, 'redirect_urls'])
```

```python
df.head()
```

```python
from ast import literal_eval
```

```python
literal_eval(df.at[0,'redirect_urls'])
```

```python
df.dropna(subset=['final_url'], inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df = df.assign(source = 'Twitter Search')
```

```python
df
```

```python
df.to_csv('./data/urls-with-redirection/twitter-search.csv', index = False)
```
